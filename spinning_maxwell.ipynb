```python
import cv2
import os
import sys
import time
import numpy as np
import random

# Caminho do vdeo
VIDEO_PATH = 'video.mp4'  # coloque o nome do seu arquivo aqui
FPS = 30
MAX_DURATION = None  # segundos (ou None para rodar o vdeo todo)

# Tabela de caracteres ASCII do mais escuro ao mais claro
ASCII_CHARS = "@%#*+=-:. "

def frame_to_ascii(frame, new_width=100):
    # Converte o frame para tons de cinza
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Redimensiona o frame
    height, width = gray.shape
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width * 0.55)
    resized = cv2.resize(gray, (new_width, new_height))

    # Converte pixels em caracteres ASCII
    ascii_img = []
    for row in resized:
        # Fixed the overflow error by changing the order of operations
        line = ''.join([ASCII_CHARS[int(pixel / 256.0 * len(ASCII_CHARS))] for pixel in row])
        ascii_img.append(line)
    return '\n'.join(ascii_img)

# --- Parte mostrada na imagem ---
cap = cv2.VideoCapture(VIDEO_PATH)
video_fps = cap.get(cv2.CAP_PROP_FPS) or 25
interval = max(1, int(video_fps // FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

max_frames = total_frames
if MAX_DURATION:
    max_frames = min(total_frames, int(FPS * MAX_DURATION))

ascii_frames = []
count = 0

while cap.isOpened() and len(ascii_frames) < max_frames:
    ret, frame = cap.read()
    if not ret:
        break
    if count % interval == 0:
        ascii_frames.append(frame_to_ascii(frame))
    count += 1

cap.release()

# Limpa o terminal conforme o sistema operacional
os.system('cls' if os.name == 'nt' else 'clear')

# Contagem regressiva
for i in range(5, 0, -1):
    sys.stdout.write("\033c")
    sys.stdout.write(f"Iniciando em {i}...\n")
    sys.stdout.flush()
    time.sleep(1)

delay = 1 / FPS

try:
    while True:
        for f in ascii_frames:
            sys.stdout.write("\033c")
            sys.stdout.write(f + "\n")
            sys.stdout.flush()
            time.sleep(delay)

        # Pausa de 20 segundos
        sys.stdout.write("\n\033[1;33mVdeo terminou. Pausando por 20 segundos. Pressione 's' para sortear um nmero.\033[0m\n")
        sys.stdout.flush()

        start_time = time.time()
        while time.time() - start_time < 20:
            # Check for user input (this is a simplified check and might not work directly in all environments)
            # In a true interactive terminal, you would use a library like `msvcrt` on Windows or `termios` on Linux/macOS
            # For this Colab environment, we'll simulate the wait and check for input after the pause.
            time.sleep(0.1) # Small sleep to avoid busy waiting

        # After the pause, wait for 's' key press (simulated)
        sys.stdout.write("\n\033[1;33mPressione 's' para sortear um nmero.\033[0m\n")
        sys.stdout.flush()
        # In a real application, you would capture a single key press here.
        # For this example in Colab, we'll just simulate the random number generation after the pause.
        input("Pressione Enter aps pressionar 's' no terminal (simulado): ") # This line is just to pause and wait for user interaction

        # Sortear nmero quando 's' for "pressionado"
        sorted_number = random.randint(1, 28)
        sys.stdout.write(f"\n\033[1;32mNmero sorteado: {sorted_number}\033[0m\n")
        sys.stdout.flush()

except KeyboardInterrupt:
    print("\n\033[1;32mExecuo interrompida pelo usurio.\033[0m")
```
